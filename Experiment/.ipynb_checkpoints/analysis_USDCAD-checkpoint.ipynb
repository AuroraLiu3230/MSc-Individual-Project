```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

raw = pd.read_csv('Backtest.csv')

df = raw.T
df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True, inplace=True)
df.rename

# Convert the numeric columns to float
numeric_columns = ['PnL', 'Fitness', 'TR', 'std(RR)', 'MDD', 'Win Rate', 'nTrade']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

strategy_mapping = {
    '0.002': 'STS1',
    '0.004': 'STS2',
    '0.005': 'STS3',
    '0.008': 'STS4'
}

data = df[df['Currency']=='EURNZD'].copy()

for threshold, strategy in strategy_mapping.items():
    data.loc[data['Threshold'] == threshold, 'Strategy'] = strategy

avg_dict = dict()
        
def update_strategy_dict(df, strategy_dict):
    for strategy in df['Strategy'].unique():
        sub_df = df[df['Strategy']==strategy]
        strategy_dict.update({
            strategy: [
                np.mean(sub_df['PnL']), np.mean(sub_df['Fitness']),
                np.mean(sub_df['TR']), np.mean(sub_df['std(RR)']),
                np.mean(sub_df['MDD']), np.mean(sub_df['Win Rate']),
                np.mean(sub_df['nTrade'])
            ]
        })

update_strategy_dict(data, avg_dict)

avgdf = pd.DataFrame(avg_dict)
avgdf['Index'] = ['PnL','Fitness','TR','std(RR)','MDD','Win Rate','nTrade']
avgdf.set_index('Index', inplace=True)
```

# box-plot

```python
# Define a function to analyze a given column
from scipy import stats

def analyze_column(data, column):
    strategies = data['Strategy'].unique()

    # Boxplot
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Strategy', y=column, data=data)
    plt.title(f'Boxplot of {column} by Strategy')
    plt.show()

    # ANOVA
    group_data = [data[data['Strategy'] == strategy][column].values for strategy in strategies]
    f_val, p_val = stats.f_oneway(*group_data)
    print(f"ANOVA for {column}: F={f_val}, p={p_val}")

    # Skewness and Kurtosis
    for strategy in strategies:
        skewness = stats.skew(data[data['Strategy'] == strategy][column])
        kurtosis = stats.kurtosis(data[data['Strategy'] == strategy][column])
        print(f"For {strategy}, {column}: skewness={skewness}, kurtosis={kurtosis}")
```

```python
print(f"\n=== Analysis for Fitness ===")
analyze_column(data, 'Fitness')
```

```python
print(f"\n=== Analysis for TR ===")
analyze_column(data, 'TR')
```

```python
print(f"\n=== Analysis for MDD ===")
analyze_column(data[data['Strategy']!='B&H'], 'MDD')
```

```python
print(f"\n=== Analysis for MDD ===")
analyze_column(data[data['Strategy']!='B&H'], 'Win Rate')
```

These histograms represent the distribution of the Profit and Loss (PnL) for each strategy. The y-axis shows the frequency (number of occurrences), and the x-axis shows the range of PnL values. The blue line (Kernel Density Estimate) gives an indication of the shape of the distribution.

Looking at the histogram for the MTSGA strategy, you can see that the PnL values are distributed mostly on the positive side, which is consistent with our previous finding that the MTSGA strategy has a positive average PnL. However, there is also a noticeable spread in the PnL values, indicating variability in the performance of this strategy.

From the plots, it is also visible that none of the strategies have a perfectly normal distribution of PnL values. This is especially important as our previous statistical tests (ANOVA and Tukey's HSD) assumed normality. The fact that our data do not perfectly meet this assumption doesn't invalidate our results, but it does mean we should be cautious in our interpretation.

The assumptions of normality and equal variances are important for ANOVA. Violations of these assumptions can lead to inaccurate results. However, ANOVA is considered to be robust to violations of these assumptions under certain conditions.

Normality: ANOVA is robust to violations of normality as long as the distributions are symmetric and not heavily-tailed (no extreme outliers), especially when the sample sizes are equal or nearly equal across groups.
Equal Variances (Homogeneity of Variances): This is the more important assumption. Violations of homogeneity of variances can lead to inaccurate results. However, when group sizes are equal, ANOVA is relatively robust to this violation as well.
In our case, we observed that none of the strategies have a perfectly normal distribution of PnL values, but the Shapiro-Wilk test did not provide strong evidence against normality. Also, Levene's test did not provide evidence against the assumption of homogeneity of variances.

Therefore, it was not inappropriate to use ANOVA in this context, but the results should be interpreted with caution. Furthermore, it's important to note that we performed ANOVA and Tukey's HSD test at a significance level of 0.1, which is higher than the conventional 0.05 level, increasing the chance of Type I errors.

However, if the assumptions of ANOVA are strongly violated, non-parametric alternatives to ANOVA, such as the Kruskal-Wallis test, can be used. These tests do not require the assumption of normality or homogeneity of variances, but they may have less statistical power (ability to detect a true effect) than ANOVA when the assumptions of ANOVA are met.




## Kruskal-Wallis

```python
# Perform Kruskal-Wallis test
kruskal_result = stats.kruskal(*[data[data['Strategy'] == strategy]['PnL'] for strategy in unique_strategies])

kruskal_result
```

The Kruskal-Wallis test returned a test statistic of approximately 15.29 and a p-value of approximately 0.054.

The Kruskal-Wallis test is a non-parametric alternative to ANOVA that tests the null hypothesis that all groups come from the same population (or different populations with the same distribution).

The p-value is greater than 0.05 but less than 0.1. This means that if we set our significance level to 0.1, we would reject the null hypothesis and conclude that there is a statistically significant difference in the PnL distributions between at least two of the strategies. However, similar to ANOVA, the Kruskal-Wallis test does not tell us which specific strategies are different.

If we want to conduct post-hoc comparisons between pairs of strategies, we can use the Dunn's test, which is a non-parametric equivalent to Tukey's HSD test.

## M-W U Test

The Mann-Whitney U test is a non-parametric test that can be used to compare two independent samples. However, please note that performing multiple tests increases the likelihood of Type I errors (false positives), so a correction method such as the Bonferroni correction is usually applied to control the family-wise error rate.

```python
from scipy.stats import mannwhitneyu
from itertools import combinations

# Get all combinations of strategies
strategy_combinations = list(combinations(unique_strategies, 2))

# Initialize an empty DataFrame to store the results
mann_whitney_results = pd.DataFrame(columns=['Group1', 'Group2', 'Statistic', 'p-value'])

# Perform Mann-Whitney U test for each pair of strategies
for group1, group2 in strategy_combinations:
    stat, p = mannwhitneyu(data[data['Strategy'] == group1]['PnL'], 
                           data[data['Strategy'] == group2]['PnL'],
                           alternative='two-sided')
    temp_df = pd.DataFrame({'Group1': [group1], 
                            'Group2': [group2], 
                            'Statistic': [stat], 
                            'p-value': [p]})
    mann_whitney_results = pd.concat([mann_whitney_results, temp_df])

# Apply Bonferroni correction
mann_whitney_results['p-value'] *= len(strategy_combinations)
mann_whitney_results['p-value'] = mann_whitney_results['p-value'].clip(upper=1)

# Display the results
mann_whitney_results.sort_values(by='p-value')
```

The pairs of strategies that show the smallest (but still not statistically significant at the 0.1 level) p-values are:

* STS1 vs. MTSGA: p-value = 0.18

* STS1 vs. MTS: p-value = 0.22

* STS1 vs. STS2: p-value = 0.94

For all other pairs, including all pairs involving the MTSGA strategy, the differences in the PnL distributions are not statistically significant at the 0.1 level after correcting for multiple testing.

These results suggest that, according to the Mann-Whitney U tests, there is no statistically significant evidence at the 0.1 level to suggest that the MTSGA strategy performs differently from any of the other strategies in terms of PnL. However, as always, these results are based on historical data, and future performance may vary.

# Total Return (TR)

```python
# Check if the assumptions of normality and homogeneity of variances are met for 'TR'

# Normality
print("Shapiro-Wilk Test for Normality:")
for strategy in unique_strategies:
    _, p = stats.shapiro(data[data['Strategy'] == strategy]['TR'])
    print(f"{strategy}: p-value = {p}")

# Homogeneity of variances
_, p = stats.levene(*[data[data['Strategy'] == strategy]['TR'] for strategy in unique_strategies])
print(f"\nLevene's Test for Homogeneity of Variances: p-value = {p}")
```

### Discussion

Levene's Test for Homogeneity of Variances:

p-value = 0.09
The p-values from the Shapiro-Wilk test are all above 0.05, which suggests that we do not reject the null hypothesis of normality for the 'TR' values of each strategy at the 0.05 significance level.

The p-value from Levene's test is also above 0.05, which suggests that we do not reject the null hypothesis of equal variances for the 'TR' values of each strategy at the 0.05 significance level.

Since both assumptions are met at the 0.05 level, it is appropriate to perform ANOVA for the 'TR' values.

## ANOVA

```python
# Perform ANOVA for 'TR'
anova_result = stats.f_oneway(*[data[data['Strategy'] == strategy]['TR'] for strategy in unique_strategies])

anova_result
```

The one-way ANOVA test for 'TR' values across different strategies returns a test statistic of approximately 1.79 and a p-value of approximately 0.087.

The null hypothesis for ANOVA is that the means of 'TR' for all strategies are equal. Given that the p-value is less than 0.1 but greater than 0.05, if we set our significance level to 0.1, we would reject the null hypothesis and conclude that there is a statistically significant difference in the mean 'TR' between at least two of the strategies.

Let's now perform post-hoc analysis to identify which specific strategies have different 'TR'. We'll use Tukey's HSD test for this.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey's HSD test for 'TR'
tukey_result = pairwise_tukeyhsd(endog=data['TR'], groups=data['Strategy'], alpha=0.1)

# Convert the result to a DataFrame
tukey_df = pd.DataFrame(data=tukey_result._results_table.data[1:], columns=tukey_result._results_table.data[0])

# Display the results
tukey_df.sort_values(by='p-adj')
```

```python
tukey_df[tukey_df['reject']==True]
```

Here are the results from Tukey's HSD post-hoc test. It compares every pair of strategies and indicates whether there's a statistically significant difference in the 'TR' between them.

The test result shows that only one pair, MTSGA and STS1, has a p-value less than 0.1 (p-value = 0.088). For this pair, the mean difference in 'TR' is -0.031, and we reject the null hypothesis of equal means at the 0.1 significance level. This suggests that the MTSGA and STS1 strategies have significantly different 'TR' values, with MTSGA having on average lower 'TR' than STS1.

For all other pairs, the differences in 'TR' are not statistically significant at the 0.1 level.

# MDD

```python
# Check if the assumptions of normality and homogeneity of variances are met for 'MDD'

# Normality
print("Shapiro-Wilk Test for Normality:")
for strategy in unique_strategies:
    _, p = stats.shapiro(data[data['Strategy'] == strategy]['MDD'])
    print(f"{strategy}: p-value = {p}")

# Homogeneity of variances
_, p = stats.levene(*[data[data['Strategy'] == strategy]['MDD'] for strategy in unique_strategies])
print(f"\nLevene's Test for Homogeneity of Variances: p-value = {p}")
```

### Discussion
The p-values from the Shapiro-Wilk test are above 0.05 for most strategies, which suggests that we do not reject the null hypothesis of normality for the 'MDD' values of these strategies at the 0.05 significance level. However, for STS2, STS3, and STS4, the p-values are below 0.05, which suggests that we reject the null hypothesis of normality for these strategies.

The p-value from Levene's test is well below 0.05, which suggests that we reject the null hypothesis of equal variances for the 'MDD' values of each strategy at the 0.05 significance level.

Given that the assumptions for ANOVA are not met, we should not perform ANOVA for the 'MDD' values. Instead, we can use the Kruskal-Wallis test, which is a non-parametric alternative to ANOVA that does not require the assumptions of normality and equal variances.

```python
# Perform Kruskal-Wallis test for 'MDD'
kruskal_result = stats.kruskal(*[data[data['Strategy'] == strategy]['MDD'] for strategy in unique_strategies])

kruskal_result
```

The Kruskal-Wallis test for 'MDD' values across different strategies returns a test statistic of approximately 69.94 and a p-value of approximately $5.06×10^{-12}$.

The null hypothesis for the Kruskal-Wallis test is that all groups come from the same population (or different populations with the same distribution). Given that the p-value is significantly less than 0.1, we reject the null hypothesis and conclude that there is a statistically significant difference in the 'MDD' distributions between at least two of the strategies.

To identify which specific strategies have different 'MDD', we can perform post-hoc pairwise comparisons. However, as the Dunn's test is not available in this environment, we can perform pairwise Mann-Whitney U tests instead.

```python
# Initialize an empty DataFrame to store the results
mann_whitney_results = pd.DataFrame(columns=['Group1', 'Group2', 'Statistic', 'p-value'])

# Perform Mann-Whitney U test for each pair of strategies
for group1, group2 in strategy_combinations:
    stat, p = mannwhitneyu(data[data['Strategy'] == group1]['MDD'], 
                           data[data['Strategy'] == group2]['MDD'],
                           alternative='two-sided')
    temp_df = pd.DataFrame({'Group1': [group1], 
                            'Group2': [group2], 
                            'Statistic': [stat], 
                            'p-value': [p]})
    mann_whitney_results = pd.concat([mann_whitney_results, temp_df])

# Apply Bonferroni correction
mann_whitney_results['p-value'] *= len(strategy_combinations)
mann_whitney_results['p-value'] = mann_whitney_results['p-value'].clip(upper=1)

# Display the results
mann_whitney_results.sort_values(by='p-value')
```

```python
mann_whitney_results[mann_whitney_results['p-value']<0.1]
```

For all other pairs, including all pairs involving the MTSGA strategy (except for B&H, RSI), the differences in 'MDD' are not statistically significant at the 0.1 level after correcting for multiple testing.

These results suggest that, according to the Mann-Whitney U tests, there are statistically significant differences in 'MDD' between some of the strategies. In particular, the MTSGA strategy has a significantly different 'MDD' compared to the B&H and RSI strategies. However, the specific context and objectives of the trading strategies should be considered when interpreting these results.
